Task 2: Unemployment Analysis with Python

Unemployment is measured by the unemployment rate which is the number of people who are unemployed as a percentage of the total labour force. We have seen a sharp increase in the unemployment rate during Covid-19, so analyzing the unemployment rate in a simplest way



IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import seaborn as sea
import matplotlib.pyplot as plt
import plotly .express as px
```

LOADING THE DATASET

```python
unemp=pd.read_csv('/content/Unemployment_Rate_upto_11_2020.csv')
unemp.head()
```

DATASET INFORMATION

```python
unemp.info()
```

STATISTICAL FEATURES

```python
unemp.describe()
```

Checking for any NULL values in the dataset

```python
unemp.isnull().sum()
```

RENAMING THE COLUMNS

```python
unemp.columns= ['States', 'Date', 'Frequency','Estimate Jobless Rate','Estimate Employed',
 'Estimate Labour Participation Rate',  'Region', 'longitude', 'latitude']
unemp.head()
```

Examine the relationship between the characteristics of the data by checking the correlation by plotting correlation heatmap

```python
plt.figure(figsize=(10,5))
sea.set(style="whitegrid")

corrmatrix=unemp.corr(numeric_only="True")
sea.heatmap(corrmatrix, annot=True, cmap="PuBu")

plt.title("Correlation Matrix")
plt.show()
```

Let's now analyze the unemployment rate by visualizing the data by looking at the estimated number of employees by India's various regions.

```python
unemp.columns= ['States', 'Date', 'Frequency','Estimate Jobless Rate','Estimate Employed',
 'Estimate Labour Participation Rate',  'Region', 'longitude', 'latitude']
plt.title("Indian Unemployment")
sea.histplot(x="Estimate Employed", hue="Region", data=unemp)
plt.show()
```

Examining the unemployment rate in India:

```python
plt.figure(figsize=(22, 20))
plt.title("Unemployment Rate")
sea.histplot(x="Estimate Jobless Rate", hue="Region", data=unemp)
plt.show()
```

Now, we will analyze unemployment rate of each INDIAN STATE by creating dashboards

```python
#creating sunburst
unemployment = unemp[['States', 'Region', 'Estimate Jobless Rate']]
figure = px.sunburst(unemployment, path=['Region', 'States'],
                     values='Estimate Jobless Rate',
                     width=900, height=900, color_continuous_scale='RdY2Gn',
                     title='Jobless Rate in India')
figure.show()
```
